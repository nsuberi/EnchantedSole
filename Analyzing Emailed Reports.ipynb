```python
import os
import pandas as pd
```

```python
REPORT_LOCATION = '~/Downloads/'
all_files = [os.path.join(root, file) for (root, dirs, files) in os.walk(os.path.expanduser(REPORT_LOCATION)) for file in files]
all_files
```

```python
DATE_FORMAT = '%m-%d-%Y'
```

```python
INVENTORY_NAMEBASE = "Sole Comfort and Enchanted Art inventory as of "

def check_name(name, matches):
    if matches in name:
        return True
    return False

inventory_files = list(filter(lambda name: check_name(name, INVENTORY_NAMEBASE), all_files))
inventory_files
```

```python
df = pd.read_csv(inventory_files[0])
df = df[df['Vend Code']=='DAN']
df = df[df['Desc1'].str.contains('PRO')]
df = df.groupby(['Desc1', 'Attr']).sum().reset_index()
df = df[df['Qty'] < 4]

def process(listOlists):
    if len(listOlists) == 2:
        return listOlists[0].fillna('') + listOlists[1].fillna('')
    else:
        return listOlists[-1].fillna('') + process(listOlists[:-1])
    
keys = pd.DataFrame({'Key_x':process([df['Desc1'], df['Attr']])})
df = pd.read_csv(inventory_files[0])
df['Key_y'] = process([df['Desc1'], df['Attr']])
data = keys.merge(df, how='inner', right_on='Key_y', left_on = 'Key_x')
data.drop(['Key_x', 'Key_y'], axis=1).sort_values(by=['Desc1', 'Attr', 'Size'])
```

```python
df = pd.read_csv(inventory_files[0])
df = df.groupby('Vend Code').sum().reset_index()
df['Avg Wholesale'] = df['Cost']/df['Qty']
df['Avg Retail'] = df['Price']/df['Qty']
df.sort_values(by='Avg Wholesale')
```
